```python
##waterjug problem -- using traditional approach

class Waterjug:
    def __init__(self,bjmax,sjmax,bj,sj,goal):
        self.bjmax=bjmax
        self.sjmax=sjmax
        self.bj=bj
        self.sj=sj
        self.goal=goal

    def fillbj(self):
        self.bj=self.bjmax
        print("(",self.bj,",",self.sj,")")

    def fillsj(self):
        self.sj=self.sjmax
        print("(",self.j,",",self.sj,")")

    def emptybj(self):
        self.bj=0
        print("(",self.bj,",",self.sj,")")

    def emptysj(self):
        self.sj=0
        print("(",self.bj,",",self.sj,")")

    def transferbjtosj(self):
        while True:
            self.bj=self.bj-1
            self.sj=self.sj+1
            if self.bj==0 or self.sj==self.sjmax:
                break
        print("(",self.bj,",",self.sj,")")

    def measure_goal(self):
        print("(",self.bj,",",self.sj,")")
        while True:
            if self.bj==self.goal or self.sj==self.goal:
                print("Successful measuring")
                break
            if self.bj==0:
                self.fillbj()
            elif self.bj>0 and self.sj<self.sjmax:
                self.transferbjtosj()
            elif self.bj>0 and self.sj==self.sjmax:
                self.emptysj()

if __name__ =='__main__':
    waterjug=Waterjug(5,3,0,0,1)
    waterjug.measure_goal()
```

```python
##waterjug problem -- using DFS approach

class Waterjug:  
    def __init__(self,bjmax,sjmax,bj,sj,goal):
        self.bjmax=bjmax
        self.sjmax=sjmax
        self.bj=bj
        self.sj=sj
        self.goal=goal
        
    def dfs(self):
        open=[]
        close=[]
        open.append((self.bj,self.sj))
        close.append((self.bj,self.sj))
        while open:
            node=open.pop()
            print(node)
            succ=[]
            bj=node[0]
            sj=node[1]
            
            #goal case
            if bj== self.goal or sj==self.goal:
                print('Successful measuring')
                return
            
            #rule 1 fill bj
            if bj==0 and (self.bjmax,sj) not in close:
                succ.append((self.bjmax,sj))
                
            #rule 2 fill sj
            if sj==0 and (bj,self.sjmax) not in close:
                succ.append((bj,self.sjmax))
                
            #rule 3 empty bj
            if bj>0 and (0,sj) not in close:
                succ.append((0,sj))
                
            #rule 4 empty sj
            if sj>0 and (bj,0) not in close:
                succ.append((bj,0))
                
            #rule 5 transfer bj to sj
            if bj>0 and sj<self.sjmax:
                if bj>=(self.sjmax-sj):
                    x=bj-(self.sjmax-sj)
                    y=self.sjmax
                else:
                    x=0
                    y=bj+sj
                if (x,y) not in close:
                    succ.append((x,y))
                    
            for i in succ[::-1]:
                open.append(i)
                close.append(i)


if __name__ =='__main__':
    big = int(input('enter big jug capacity value : '))
    small = int(input('enter small jug capacity value : '))
    goal = int(input('enter goal capacity value : '))
    waterjug=Waterjug(big,small,0,0,goal)
    waterjug.dfs()
```
